```python
# Drop the 'license' column due to excessive null values and low relevance
if 'license' in df.columns:
    df.drop('license', axis=1, inplace=True)
    print("\nDropped 'license' column.")
```

```python
# Clean 'price' and 'service fee' columns
# Remove '$', ',', handle non-numeric values, and convert to float
for col in ['price', 'service fee']:
    if df[col].dtype == 'object':
        df[col] = df[col].replace({'\$': '', ',': ''}, regex=True)
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Fill null values
# For reviews per month, null likely means no reviews, so 0 is appropriate.
df['reviews per month'].fillna(0, inplace=True)
```

```python
# For last review date, we'll leave NaT for now as it indicates no review history
# For review rate number, fill with the median
median_review_rate = df['review rate number'].median()
df['review rate number'].fillna(median_review_rate, inplace=True)
```

```python
# For construction year, fill with the mode (most common year)
mode_construction_year = df['Construction year'].mode()[0]
df['Construction year'].fillna(mode_construction_year, inplace=True)
```

```python
# Drop rows where essential data like price or host name is missing
df.dropna(subset=['price', 'host name', 'host_identity_verified'], inplace=True)
```

Visulization & Insights

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Group by neighbourhood group
counts = df.groupby('neighbourhood group').size().sort_values(ascending=False)
avg_price = df.groupby('neighbourhood group')['price'].mean().sort_values(ascending=False)

print(counts.head())
print(avg_price.head())

# which neighbourhood group has the highets number of listings?
plt.figure(figsize=(7,4))
sns.barplot(x=counts.index, y=counts.values, palette='Blues_d')
plt.title('Listings by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Number of Listings')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# Distribution of Listings Across Boroughs
plt.figure()
sns.countplot(y=df['neighbourhood group'], order=df['neighbourhood group'].value_counts().index, palette='viridis')
plt.title('Number of Airbnb Listings per Borough')
plt.xlabel('Number of Listings')
plt.ylabel('Borough')
plt.tight_layout()
```

```python
# Most Popular Room Types
plt.figure()
room_type_counts = df['room type'].value_counts()
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('plasma'))
plt.title('Distribution of Room Types')
plt.ylabel('')
```

```python
# Price Distribution by Borough
plt.figure()
# Filtering out extreme outliers for better visualization
price_cap = df['price'].quantile(0.95)
sns.boxplot(data=df[df['price'] < price_cap], x='price', y='neighbourhood group', palette='magma')
plt.title('Price Distribution by Borough (Outliers Removed)')
plt.xlabel('Price ($)')
plt.ylabel('Borough')
plt.tight_layout()
```

```python

# Price vs service fee correlation
corr_val = df[['price','service fee']].corr().iloc[0,1]
print(corr_val)
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='price', y='service fee', alpha=0.3)
sns.regplot(data=df, x='price', y='service fee', scatter=False, color='red')
plt.title('Service Fee vs Price (corr = ' + str(round(corr_val, 3)) + ')')
plt.tight_layout()
plt.show()
```

```python
# Host Verification Status
plt.figure()
sns.countplot(x=df['host_identity_verified'], palette='coolwarm')
plt.title('Host Identity Verification Status')
plt.xlabel('Is Host Identity Verified?')
plt.ylabel('Count of Hosts')
plt.tight_layout()
```

```python

# Avg review rate number by neighbourhood group and room type
pivot_rt = df.pivot_table(index='neighbourhood group', columns='room type', values='review rate number', aggfunc='mean')
print(pivot_rt.head())
plt.figure(figsize=(7,4))
sns.heatmap(pivot_rt, annot=True, fmt='.2f', cmap='YlGnBu')
plt.title('Avg Review Rate by Neighbourhood Group and Room Type')
plt.tight_layout()
plt.show()
```

```python

# Verified identity vs positive reviews (>=4)
import numpy as np
if 'host_identity_verified' in df.columns:
    df['positive_review'] = np.where(df['review rate number'] >= 4, 1, 0)
    id_rate = df.groupby('host_identity_verified')['positive_review'].mean().sort_values(ascending=False)
    print(id_rate)
    plt.figure(figsize=(5,4))
    sns.barplot(x=id_rate.index.astype(str), y=id_rate.values, palette='Set2')
    plt.title('Share of Positive Reviews by Host Identity Verification')
    plt.xlabel('Host Identity Verified')
    plt.ylabel('Share Positive Reviews')
    plt.ylim(0,1)
    plt.tight_layout()
    plt.show()
```

```python
# Impact of Cancellation Policy on Price
plt.figure()
sns.boxplot(data=df[df['price'] < price_cap], x='price', y='cancellation_policy', palette='crest')
plt.title('Price Distribution by Cancellation Policy')
plt.xlabel('Price ($)')
plt.ylabel('Cancellation Policy')
plt.tight_layout()
```

```python
# Top 10 Most Active Neighbourhoods
plt.figure()
top_10_neighbourhoods = df['neighbourhood'].value_counts().nlargest(10)
sns.barplot(y=top_10_neighbourhoods.index, x=top_10_neighbourhoods.values, palette='rocket')
plt.title('Top 10 Neighbourhoods by Number of Listings')
plt.xlabel('Number of Listings')
plt.ylabel('Neighbourhood')
plt.tight_layout()
```

```python
# Top 10 hosts by calculated host listings count
host_counts = df.groupby('host name')['calculated host listings count'].max().sort_values(ascending=False).head(10)
print(host_counts)
plt.figure(figsize=(7,4))
sns.barplot(x=host_counts.values, y=host_counts.index, palette='mako')
plt.title('Top 10 Hosts by Calculated Host Listings Count')
plt.xlabel('Listings Count')
plt.ylabel('Host Name')
plt.tight_layout()
plt.show()
```

```python
# Listing Availability Distribution
plt.figure()
sns.histplot(df['availability 365'], bins=30, kde=True, color='purple')
plt.title('Distribution of Listing Availability')
plt.xlabel('Availability (Days per Year)')
plt.ylabel('Number of Listings')
plt.tight_layout()
```

```python

# Host listings count vs availability
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='calculated host listings count', y='availability 365', alpha=0.3)
sns.regplot(data=df, x='calculated host listings count', y='availability 365', scatter=False, color='red')
plt.title('Availability vs Host Listings Count')
plt.tight_layout()
plt.show()
```

```python
# Geospatial Distribution of Listings
plt.figure(figsize=(10, 10))
sns.scatterplot(data=df, x='long', y='lat', hue='neighbourhood group', s=5, alpha=0.5)
plt.title('Geospatial Map of Airbnb Listings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Borough')
plt.tight_layout()
```

```python
# Trend of Reviews Over Time
plt.figure()
# We use 'last review' year for this trend analysis
df['review_year'] = df['last review'].dt.year
review_trend = df['review_year'].value_counts().sort_index()
# Filter for plausible years
review_trend = review_trend[(review_trend.index >= 2012) & (review_trend.index <= 2022)]
review_trend.plot(kind='line', marker='o', color='teal')
plt.title('Trend of Last Reviews by Year')
plt.xlabel('Year')
plt.ylabel('Number of Reviews Logged')
plt.grid(True)
plt.tight_layout()
```
